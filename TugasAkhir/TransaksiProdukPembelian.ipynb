```python
ewallet_purchases = df[df['Metode Pembayaran'] == 'E-Wallet']['Barang Kebutuhan']
print("Produk yang dibeli menggunakan E-wallet:")
print(ewallet_purchases)
```

```python
purchase_dates = df['Tanggal Transaksi']
print("Tanggal pembelian:")
print(purchase_dates)
```

```python
sabun_purchases = df[df['Barang Kebutuhan'] == 'Sabun']['Tanggal Transaksi']
print("Tanggal pembelian produk Sabun:")
print(sabun_purchases)
```

```python
pasta_gigi_purchases = df[df['Barang Kebutuhan'] == 'Pasta gigi']
jumlah_pasta_gigi_purchases = len(pasta_gigi_purchases)
print("Jumlah pembelian produk Pasta gigi:", jumlah_pasta_gigi_purchases)
```

```python
import matplotlib.pyplot as plt
```

```python
jumlah_pembelian_produk = df['Barang Kebutuhan'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(jumlah_pembelian_produk.index, jumlah_pembelian_produk.values)
plt.xlabel('Produk')
plt.ylabel('Jumlah Pembelian')
plt.title('Jumlah Pembelian Setiap Produk')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Tanggal Transaksi'] = pd.to_datetime(df['Tanggal Transaksi'], dayfirst=True)
```

```python
jumlah_pembelian_tanggal = df.groupby('Tanggal Transaksi').size()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(jumlah_pembelian_tanggal.index, jumlah_pembelian_tanggal.values)
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Pembelian')
plt.title('Pembelian Produk per Tanggal')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
jumlah_pembelian_pembayaran = df['Metode Pembayaran'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(jumlah_pembelian_pembayaran.index, jumlah_pembelian_pembayaran.values)
plt.xlabel('Metode Pembayaran')
plt.ylabel('Jumlah Pembelian')
plt.title('Pembelian Produk per Metode Pembayaran')
plt.show()
```
